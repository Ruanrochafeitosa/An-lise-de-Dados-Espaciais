```python
import pandas as pd
import geopandas as geo
import folium

#Lendo o arquivo CSV
#Contendo  a distribuição global de vulcões em 2010
df = pd.read_csv("volcano_data_2010.csv")
df
```

```python
#Aqui a gente filtra os dados que a gente quer
df = df[["Year", "Name", "Country", "Latitude", "Longitude", "Type"]]
df
```

```python
#Temos que unir as variáveis de longitude e latitude para que ela se torne um geomtre um geodataframe, no final ela vai virar 
#uma variável chamada geometry e aparece como coluna nova
geometrias = geo.points_from_xy(df["Longitude"], df["Latitude"])
geo_df = geo.GeoDataFrame(df, geometry=geometrias)
geo_df

```

```python
import folium.plugins


mapa = folium.Map(location=[15, 30], tiles="Cartodb dark_matter", zoom_start=2)
# Isso serve para que possamos fazer um mapa e com isso definimos uma localização aleatória e em tiles é o tema do gráfico
#e o zoom é para definir o zoom inicial, ou seja vamos ter uma visão mais ampla.
mapa

heatmap_dados = [[point.xy[1][0], point[0][0]] for point in geo_df["geometry"]]
#Essa é uma forma de criar umalista sem precisar do for que nem o exemplo lá de cima

#vamos criar um heatmap para visualizar  a densidade de vulcões
folium.plugins.HeatMap(heatmap_dados).add_to(mapa)
#aqui chamamos a bibliotaca folium, depois o módulo plugins e ai passamos o mapa de calor e adicioamos o parâmetro heatmap_dados
#depois adicionamos tudo no mapa com add_to(mapa)
mapa
```
